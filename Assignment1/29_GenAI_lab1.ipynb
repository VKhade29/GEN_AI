```python
# ---- Semantic Search with LangChain + Cohere ----

from langchain_cohere import CohereEmbeddings
from langchain_community.vectorstores import FAISS

documents = [
    "Convolutional Neural Networks are widely used for image classification tasks",
    "Recurrent Neural Networks are designed for sequence and time-series data",
    "Support Vector Machines are effective for high-dimensional datasets",
    "K-Means is an unsupervised clustering algorithm"
]

# Create embeddings
embeddings = CohereEmbeddings(
    model="embed-english-v3.0"
)

# Build vector store
vectorstore = FAISS.from_texts(documents, embeddings)

# Query
query = "Which algorithm is used for image processing?"

# Semantic search
results = vectorstore.similarity_search(query, k=2)

# Output
for i, doc in enumerate(results):
    print(f"Result {i+1}: {doc.page_content}")
```

```python
from langchain_cohere import ChatCohere
from langchain_core.messages import HumanMessage, SystemMessage



llm = ChatCohere(
    model="command-r7b-12-2024",
    temperature=0
)

messages = [
    SystemMessage(
        content=(
            "You are a sentiment analysis engine. Analyze the provided text and classify its sentiment strictly as 'Positive', 'Negative', or 'Neutral'."
        )
    ),
    HumanMessage(
        content="The product quality is amazing but the delivery was very slow."
    )
]

result = llm.invoke(messages)

print(result.content)
```
